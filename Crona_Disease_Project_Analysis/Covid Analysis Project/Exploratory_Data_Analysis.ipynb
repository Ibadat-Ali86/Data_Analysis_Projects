**clean_missing_values_by_priority()**

Purpose: Handle missing values strategically based on business importance.

## Summary Points:

- Converts month_start_date to datetime for time-based operations.

- Core Metrics (Population, GDP, HDI, Cases, Deaths):

- Demographics  Forward fill + median.

- COVID metrics  Replace early pandemic NaNs (before March 2020) with 0 and forward fill later values.

## Vaccination Data:

- Pre-2021 values replaced with 0 (pre-vaccine era).

- Post-2021 values interpolated per country.

- Testing & Policy Data:

- Forward filled by country.

-Default assumptions applied (e.g., 10% positivity rate, median stringency).

- Healthcare System Columns:

- Missing data flagged with new _available columns.

- Missing values replaced with 0 for consistency.

- Returns a clean dataset ready for validation.

```python
def clean_missing_values_by_priority(final_covid):
    """Strategic cleaning based on business importance"""
   
    print(" Strategic Missing Value Cleaning...")
    final_covid_clean = final_covid.copy()
    final_covid_clean['month_start_date'] = pd.to_datetime(final_covid_clean['month_start_date'])
   
    #  Core Business Metrics (Demographics + COVID) 
    core_cols = [
        'total_population', 'human_development_index_score', 'gdp_per_capita_usd',
        'total_confirmed_cases_month_end', 'monthly_new_cases',
        'total_confirmed_deaths_month_end', 'monthly_new_deaths'
    ]
   
    for col in core_cols:
        if col in final_covid_clean.columns:
            if 'population' in col or 'gdp' in col or 'human_development' in col:
                # Demographics
                final_covid_clean[col] = final_covid_clean.groupby('country_name')[col].ffill()
                final_covid_clean[col] = final_covid_clean[col].fillna(final_covid_clean[col].median())
            else:
                #  early pandemic = 0, then forward fill
                early_mask = final_covid_clean['month_start_date'] < '2020-03-01'
                final_covid_clean.loc[early_mask, col] = final_covid_clean.loc[early_mask, col].fillna(0)
                final_covid_clean[col] = final_covid_clean.groupby('country_name')[col].ffill()
                final_covid_clean[col] = final_covid_clean[col].fillna(0)
   
    #  Vaccination Data (Timeline-Aware) 
    vaccination_cols = [col for col in final_covid_clean.columns if 'vaccin' in col or 'vaccine' in col]
   
    for col in vaccination_cols:
        # Pre-vaccine era = 0
        pre_vaccine = final_covid_clean['month_start_date'] < '2021-01-01'
        final_covid_clean.loc[pre_vaccine, col] = 0
       
        # Vaccine era
        vaccine_era = final_covid_clean['month_start_date'] >= '2021-01-01'
        if vaccine_era.sum() > 0:
            #  Using transform with interpolate for proper groupby application
            final_covid_clean.loc[vaccine_era, col] = (
                final_covid_clean.loc[vaccine_era]
                .groupby('country_name')[col]
                .transform(lambda x: x.interpolate(method='linear', limit_direction='both'))
            )
       
        final_covid_clean[col] = final_covid_clean[col].fillna(0)
   
    # Testing & Policy Data 
    testing_policy_cols = [col for col in final_covid_clean.columns if 'test' in col or 'stringency' in col]
   
    for col in testing_policy_cols:
        final_covid_clean[col] = final_covid_clean.groupby('country_name')[col].ffill()
       
        if 'positivity' in col:
            final_covid_clean[col] = final_covid_clean[col].fillna(10.0)  # Conservative 10%
        elif 'stringency' in col:
            final_covid_clean[col] = final_covid_clean[col].fillna(final_covid_clean[col].median())
            final_covid_clean[col] = final_covid_clean[col].clip(0, 100)
        else:
            final_covid_clean[col] = final_covid_clean[col].fillna(0)
   
    # Healthcare System (High Missing - Flag Approach) 
    healthcare_cols = [col for col in final_covid_clean.columns if 'icu' in col or 'hosp' in col]
   
    for col in healthcare_cols:
        # Create availability flag for analysis
        final_covid_clean[f'{col}_available'] = (~final_covid_clean[col].isnull()).astype(int)
        final_covid_clean[col] = final_covid_clean[col].fillna(0)
   
    print(" Strategic cleaning completed!")
    return final_covid_clean
```

### **Step 3: Business Logic Validation**
The business logic validation step involves checking whether the input data meets the requirements of the business rules. In this case, we will check whether the input data contains any missing values or outliers. If there are any issues, we will return an error message indicating what needs to be fixed.

```python
def validate_business_logic(final_covid):
    """Validate and fix business rule violations"""
    
    print(" Business Logic Validation...")
    final_covid_valid = final_covid.copy()
    fixes = 0
    
    #  Deaths  Cases
    if all(c in final_covid_valid.columns for c in ['total_confirmed_deaths_month_end', 'total_confirmed_cases_month_end']):
        deaths_exceed = final_covid_valid['total_confirmed_deaths_month_end'] > final_covid_valid['total_confirmed_cases_month_end']
        if deaths_exceed.sum() > 0:
            final_covid_valid.loc[deaths_exceed, 'total_confirmed_deaths_month_end'] = \
                final_covid_valid.loc[deaths_exceed, 'total_confirmed_cases_month_end']
            fixes += deaths_exceed.sum()
            print(f"    Fixed {deaths_exceed.sum()} cases where deaths > cases")
    
    #  Percentages 0-100%
    percentage_cols = [col for col in final_covid_valid.columns if 
                      any(x in col for x in ['percent', 'per_100', 'positivity', 'fatality_rate'])]
    
    for col in percentage_cols:
        invalid = (final_covid_valid[col] > 100) | (final_covid_valid[col] < 0)
        if invalid.sum() > 0:
            final_covid_valid[col] = final_covid_valid[col].clip(0, 100)
            fixes += invalid.sum()
            print(f"    Fixed {invalid.sum()} invalid percentages in {col}")
    
    #  No negative counts
    count_cols = [col for col in final_covid_valid.columns if 
                 any(x in col for x in ['total_', 'monthly_', 'avg_', 'peak_'])]
    
    for col in count_cols:
        if final_covid_valid[col].dtype in ['int64', 'float64', 'Int64', 'float32']:
            negative = final_covid_valid[col] < 0
            if negative.sum() > 0:
                final_covid_valid.loc[negative, col] = 0
                fixes += negative.sum()
    
    print(f" Validation completed! Fixed {fixes} violations")
    return final_covid_valid
```

## Summary Points:

Rule 1: Fix cases where deaths > cases by setting deaths = cases.

Rule 2: Constrain percentage-based columns (e.g., positivity, fatality) within 0100%.

Rule 3: Replace all negative counts with 0 (ensures no impossible negative data).

Tracks and reports total number of fixes applied.

Returns a logically consistent dataset.

### **Step 5: Monthly Feature Engineering**


```python
def create_monthly_business_features(final_covid):
    """Create features aligned with business objectives"""
    
    print(" Monthly Feature Engineering...")
    final_covid_enhanced = final_covid.copy()
    final_covid_enhanced = final_covid_enhanced.sort_values(['country_name', 'month_start_date'])
    
    # TEMPORAL FEATURES 
    final_covid_enhanced['quarter'] = final_covid_enhanced['month_start_date'].dt.quarter
    final_covid_enhanced['pandemic_year'] = final_covid_enhanced['year'] - 2020
    
    # Pandemic phases for temporal analysis
    final_covid_enhanced['pandemic_phase'] = pd.cut(
        final_covid_enhanced['pandemic_year'], 
        bins=[-1, 1, 2, 3, 10],
        labels=['Year_1', 'Year_2', 'Year_3', 'Later']
    )
    
    #  GROWTH INDICATORS 
    # Cases and vaccination growth rates
    growth_cols = ['total_confirmed_cases_month_end', 'people_vaccinated_per_100_month_end']
    for col in growth_cols:
        if col in final_covid_enhanced.columns:
            final_covid_enhanced[f'{col}_growth_rate'] = final_covid_enhanced.groupby('country_name')[col].pct_change()
    
    # POLICY EFFECTIVENESS 
    if all(c in final_covid_enhanced.columns for c in ['monthly_new_cases', 'avg_government_response_stringency']):
        final_covid_enhanced['policy_effectiveness'] = \
            final_covid_enhanced['monthly_new_cases'] / (final_covid_enhanced['avg_government_response_stringency'] + 1)
    
    # Policy consistency (within-month variation)
    if all(c in final_covid_enhanced.columns for c in ['max_government_response_stringency', 'min_government_response_stringency']):
        final_covid_enhanced['policy_consistency'] = \
            final_covid_enhanced['max_government_response_stringency'] - final_covid_enhanced['min_government_response_stringency']
    
    #  COMPARATIVE CATEGORIES 
    # Development levels for geographic comparison
    if 'human_development_index_score' in final_covid_enhanced.columns:
        final_covid_enhanced['development_level'] = pd.cut(
            final_covid_enhanced['human_development_index_score'],
            bins=[0, 0.55, 0.70, 0.80, 1.0],
            labels=['Low', 'Medium', 'High', 'Very_High']
        )
    
    # Vaccination progress levels
    if 'people_vaccinated_per_100_month_end' in final_covid_enhanced.columns:
        final_covid_enhanced['vaccination_level'] = pd.cut(
            final_covid_enhanced['people_vaccinated_per_100_month_end'],
            bins=[0, 30, 60, 80, 100],
            labels=['Low', 'Medium', 'High', 'Very_High'],
            include_lowest=True
        )
    
    print(" Feature engineering completed!")
    return final_covid_enhanced
```

## **Outlier Handling Techniques**

```python

def clean_critical_data_quality_outliers(df):
    
  
    
    print(" CLEANING CRITICAL OUTLIERS (Data Quality Issues Only)")
    print("*" * 60)
    
    df_clean = df.copy()
    total_corrections = 0
    
    #  Deaths cannot exceed cases
    if 'total_confirmed_deaths' in df_clean.columns and 'total_confirmed_cases' in df_clean.columns:
        deaths_exceed = df_clean['total_confirmed_deaths'] > df_clean['total_confirmed_cases']
        count = deaths_exceed.sum()
        if count > 0:
            print(f"Fixing {count} records where deaths > cases")
            df_clean.loc[deaths_exceed, 'total_confirmed_deaths'] = df_clean.loc[deaths_exceed, 'total_confirmed_cases']
            total_corrections += count
    
    #  Negative counts (impossible)
    negative_cols = ['total_confirmed_cases', 'total_confirmed_deaths', 'daily_new_cases', 'daily_new_deaths']
    
    for col in negative_cols:
        if col in df_clean.columns:
            negative_mask = df_clean[col] < 0
            count = negative_mask.sum()
            if count > 0:
                print(f"Fixing {count} negative values in {col}")
                df_clean.loc[negative_mask, col] = 0
                total_corrections += count
    
    # Test positivity > 100% (impossible)
    if 'test_positivity_rate' in df_clean.columns:
        invalid_positivity = df_clean['test_positivity_rate'] > 100
        count = invalid_positivity.sum()
        if count > 0:
            print(f"     Fixing {count} test positivity rates > 100%")
            df_clean.loc[invalid_positivity, 'test_positivity_rate'] = 100
            total_corrections += count
    
    #  Invalid percentages
    percentage_cols = ['percent_aged_65_plus', 'percent_aged_70_plus', 'people_vaccinated_per_100_pop']
    
    for col in percentage_cols:
        if col in df_clean.columns:
            # Fix > 100%
            over_100 = df_clean[col] > 100
            count_over = over_100.sum()
            if count_over > 0:
                print(f"   Fixing {count_over} values > 100% in {col}")
                df_clean.loc[over_100, col] = 100
                total_corrections += count_over
            
            # Fix < 0%
            under_0 = df_clean[col] < 0
            count_under = under_0.sum()
            if count_under > 0:
                print(f"     Fixing {count_under} negative values in {col}")
                df_clean.loc[under_0, col] = 0
                total_corrections += count_under
    
    # Summary
    print(f"\n CRITICAL CLEANING COMPLETE!")
    if total_corrections > 0:
        print(f" Total corrections: {total_corrections}")
        print(f" Data ready for your existing pipeline")
    else:
        print(f" No critical issues found - excellent data quality!")
    print(f" Later: Epidemiological outlier analysis in EDA stage")
    
    return df_clean
```
